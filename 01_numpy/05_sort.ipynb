# 정렬
- 배열 값을 기준에 따라 순서대로 배치하는 작업임
- 오름차순/내림차순 정렬 가능함
- 원본 유지 여부 확인 필요함


```python
import numpy as np
```

```python
# np.sort(ndarray)
# - 정렬된 새 배열을 반환
# - 원본 배열은 유지
# - 원본 데이터 유지 여부는 데이터 전처리에 있어서 중요한 부분이다.
arr = np.random.randint(1, 11, 10) #1부터 11 미만 사이의 10개
print(arr)

print("np.sort(arr)", np.sort(arr)) #정렬 (오름차순으로)
print("arr: ", arr)  #원본 유지 확인
```

```python
# 내림차순(역순 정렬) -> 슬라이싱 방식 이용
np.sort(arr) [::-1]
```

```python
# 2차원 배열 형태의 데이터 정렬
# - axis에 따라 정렬 기준이 달라짐
# - axis = None     - 전체 (1차원으로 변경 후 전체 정렬)
# - axis = 0        - 각 행 안에서 열 방향으로 정렬 (세로정렬)
# - axis = 1        - 각 열 안에서 행 방향으로 정렬 (가로정렬)

np.random.seed(1)
arr = np.random.randint(1, 15, (3,4))  #1부터 15미만의 3행4열
print(arr)

print("np.sort(arr, axis=None)")
print(np.sort(arr, axis=None))

print("np.sort(arr, axis=0)")
print(np.sort(arr, axis=0))

print("np.sort(arr, axis=1)")
print(np.sort(arr, axis=1))
```

## np.argsort()
- 배열 요소를 정렬했을 때의 인덱스를 반환함
- 정렬된 값 자체가 아니라 정렬 순서임
- 실제 정렬값은 `arr[indices]`처럼 다시 인덱싱해야 함


```python
# argsort()의 결과값은 정렬 순서를 만드는 인덱스다.

arr = np.array([40, 10, 20, 30])
indices = np.argsort(arr)
print(indices) #값을 정렬하는게 아니라 정렬된 값의 인덱스 순서를 보여줌 (0번 인덱스가 40이므로 제일 끝으로 감)
print(arr[indices])
```

```python
# 역순 정렬
arr = np.array([40, 10, 20, 30])
indices = np.argsort(-arr) # arr * -1  (마이너스기호를 붙여서 큰 값이 제일 작아지게 만들어서 앞으로 가게 만듦)
print(indices)
print(arr[indices])
```

```python
# 2차원 argsort(axis = 0)
arr = np.array([[30, 10, 20], [60, 40, 50]])  #2행3열짜리

indices = np.argsort(arr, axis=0)
print(indices)

print()

print(arr[indices])
```

```python
# 2차원 argsort(axis = 1)
arr = np.array([[30, 10, 20], [60, 40, 50]])  #2행3열짜리

indices = np.argsort(arr, axis=1)
print(indices)
```

```python
# 문제1
# ['홍길동' '유관순' '이순신' '신사임당'] 으로 출력되게 하기

# 각 배열의 같은 인덱스 요소가 한 학생의 이름, 점수를 나타내고 있음
# 학생의 이름으로 점수 내림차순 순서로 정렬
names = np.array(['홍길동', '신사임당', '이순신', '유관순'])
scores = np.array([90, 30, 50, 70])

# indices = np.argsort(names)
# print(indices)
# print(names[indices])

indices = np.argsort(-scores)
print(names[indices])
```

```python
# 문제2
# 우선순위에 따른 작업 정렬(작은 수 일수록 높은 우선 순위)
priorities = np.array([3, 1, 2, 5, 4])
tasks = np.array(["Task A", "Task B", "Task C", "Task D", "Task E"])

# 결과: ["Task B" "Task C" "Task A" "Task E" "Task D"] 나오게 하기

print(tasks[np.argsort(priorities)])
```
