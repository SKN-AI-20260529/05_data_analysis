### [실습 1] 1차원 배열 정렬


다음 1차원 배열을 오름차순 및 내림차순으로 정렬하세요.


```python
arr = np.array([50, 10, 30, 40, 20])
```

````python
import numpy as np
arr = np.array([50, 10, 30, 40, 20])
print(np.sort(arr))

print(np.sort(arr)[::-1])
````

### [실습 2] 2차원 배열 축 기준 정렬


다음 2차원 배열을 행과 열을 기준으로 각각 정렬하세요.


```python
arr2d = np.array([[8, 3, 5], [2, 9, 1]])
```

````python
arr2d = np.array([[8, 3, 5], [2, 9, 1]])

sorted_by_row = np.sort(arr2d, axis=1)
print("--- 행 기준 정렬 (axis=1) ---")
print(sorted_by_row)

sorted_by_col = np.sort(arr2d, axis=0)
print("\n--- 열 기준 정렬 (axis=0) ---")
print(sorted_by_col)
````

### [실습 3] 인덱스를 반환하는 정렬


다음 1차원 배열에서 정렬된 결과의 인덱스를 반환하고, 그 인덱스를 사용하여 배열을 정렬하세요.


```python
arr = np.array([15, 10, 20])
```

````python
arr = np.array([15, 10, 20])

# 1. 정렬된 결과의 인덱스 반환
sorted_indices = np.argsort(arr)
print("정렬된 인덱스:", sorted_indices)

# 2. 반환된 인덱스를 사용하여 배열 정렬
sorted_arr = arr[sorted_indices]
print("인덱스로 배열 정렬:", sorted_arr)
````
