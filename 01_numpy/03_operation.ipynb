## n차원 배열 연산
- 같은 shape끼리 원소별 연산 가능함
- 같은 위치의 값끼리 계산함
- 결과 shape도 기존 shape와 같음


```python
arr1 = np.arange(1,10).reshape(3,3)
print(arr1)
arr2 = np.full_like(arr1, 2)
print(arr2)

## 사칙연산 하나씩 주석 풀어서 해보기
# print(arr1+arr2)  #더하기
# print(arr1-arr2)  #빼기
# print(arr1*arr2)  #곱하기
# print(arr1/arr2)  #나누기
# print(arr1//arr2)  #몫
# print(arr1%arr2)  #나머지
# print(arr1**arr2)  #제곱
```

## broadcasting 연산 방식
- shape가 다른 배열도 규칙이 맞으면 연산 가능함
- 작은 배열을 큰 배열 shape에 맞춰 해석함
- 실제 배열을 복사해서 키우는 것은 아님

**broadcasting 규칙**
- 뒤쪽 차원부터 비교함
- 두 차원이 같으면 가능함
- 둘 중 하나가 1이면 가능함
- 둘 다 다르고 1도 아니면 불가함

![](https://d.pr/i/D0XITM+)
![](https://d.pr/i/fka9ED+)


```python
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.full((3,),2)

print("arr1\n", arr1)
print("arr2\n", arr2)

print("arr1+arr2:\n", arr1+arr2)
```

```python
arr = np.array([1,2])
print(arr.shape) #(2,)

print(arr1+arr)
# (2,3)+(1,2) -> 브로드캐스팅 불가
# 제일 끝 차원인 3,2가 같지 않고, 둘 중 1이 없음.
```

```python
# 스칼라(값1개) 또는 길이(size)가 1인 배열과의 연산
# -> 전체 원소에 스칼라 또는 배열의 값을 적용

print("arr1:\n", arr1)
print("arr1+10:\n", arr1+10)
print("arr1+[10]:\n", arr1+[10])
print("arr1+[[10]]:\n", arr1+[[10]])
print("arr1+[[[10]]]:\n", arr1+[[[10]]])
```

```python
arr3 = np.array([[1],[2]])
print("arr3\n", arr3) #arr3.shape = (2,1)
print("arr1\n", arr1) #arr1.shape = (2,3)

print(arr1+arr3)
#arr3.shape = (2,1) -> (2,3)으로 브로드캐스팅
#[[1,1,1]
#[2,2,2]]
```

```python
arr1 = np.array([[0, 0, 0], [1, 1, 1], [2, 2, 2]])
arr2 = np.array([5, 6, 7])
print(arr1,"\n")
print(arr2,"\n")
arr3 = arr1 + arr2  #[5,6,7][6,7,8][7,8,9]
#눈코딩 후 shape, 결과 확인
print(arr1+arr2)
```

```python
arr1 = np.array([[1, 2], [3, 4]])
print(arr1,"\n")
arr2 = np.array([[5], [6]]) #[5 5] [6 6] 으로 자리 만듦
print(arr2,"\n")
arr3 = arr1 + arr2 #[6 7][9 10]
#눈코딩 후 shape, 결과 확인
print(arr1+arr2,"\n") #[6 7][9 10]
```

## 행렬곱셈 Matrix Multiplication
- 점곱연산 Dot Product, 내적이라고도 함
- NumPy의 `*` 연산과 행렬곱은 다름
- `*`: 같은 위치 원소끼리 곱함
- `@`, `np.dot()`, `np.matmul()`: 행렬곱 수행함

**계산 방식**
- 왼쪽 행과 오른쪽 열을 짝지어 계산함
- 같은 위치끼리 곱한 뒤 모두 더함
- 결과의 한 칸은 왼쪽 행과 오른쪽 열의 내적값임

**shape 조건**
- 첫 번째 행렬의 열 수와 두 번째 행렬의 행 수가 같아야 함
- 행렬곱 shape: `(m, n) @ (n, p) -> (m, p)`
- 결과 shape: 바깥쪽 숫자인 `(m, p)`


- **수식**: 행렬 $ A $가 $ m \times n $ 크기이고, 행렬 $ B $가 $ n \times p $ 크기일 때, 행렬 곱 $ A \times B $의 결과는 $ m \times p $ 크기의 행렬이다. 곱셈 과정은 다음과 같다:
  $$
  (A \times B)_{ij} = \sum_{k=1}^{n} A_{ik} \times B_{kj}
  $$
  즉, $ A $의 i번째 행과 $ B $의 j번째 열에 해당하는 원소들을 곱하고, 그 결과를 더하여 새로운 행렬의 $ (i, j) $ 위치에 배치한다.

- **예시**:
  $$
  A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}, \quad B = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}
  $$
  $$
  A \times B = \begin{pmatrix} 1 \cdot 5 + 2 \cdot 7 & 1 \cdot 6 + 2 \cdot 8 \\ 3 \cdot 5 + 4 \cdot 7 & 3 \cdot 6 + 4 \cdot 8 \end{pmatrix} = \begin{pmatrix} 19 & 22 \\ 43 & 50 \end{pmatrix}
  $$

- **결과**: 새로운 **행렬**.

```python
a = [[1,2], [3,4]]
b = [[5,6], [7,8]]

a = np.array(a)
b = np.array(b)

# numpy *연산
print(a*b, "\n")

# 내적 연산(행렬 곱)
print(np.matmul(a,b), "\n")
print(np.dot(a,b), "\n")
print(a@b)
```

```python
c = np.array([[1,2]])   #(1,2) 1행 2열
d = np.array([[3],[4]]) #(2,1) 2행 1열

print("c@d= ", c@d)  #2행2열 2차원 행렬 안에 1행1열 칸에만 값이 있는 구조
```

```python
e = np.array([[1, 2, 3],
              [4, 5, 6]])    #(2,3)
f = np.array([[7, 8],
              [9, 10],
              [11, 12]])     #(3,2)
# 일단 작은 쪽으로 맞춰지므로 2행2열짜리가 나올 것임

print("e@f= ", e@f)  #(2,2)
```

## 기타 연산
- 절대값, 반올림, 올림, 내림, 버림 등 기본 수학 함수 정리임


```python
# 절댓값

arr1 = np.array([[1, -2], [3, -4]])
print(np.abs(arr1))
```

```python
# ceil, round, floor, trunc
# 올림, 반올림,   내림,   버림

arr2 = np.array([-1.234, 1.234, 2.345, 3.456, 4.567])
print("np.ceil(arr2): ", np.ceil(arr2))
print("np.round(arr2): ", np.round(arr2))

print("np.floor(arr2): ", np.floor(arr2))
print("np.trunc(arr2): ", np.trunc(arr2))
```

### axis 속성
- 연산을 처리할 축 지정값임
- `axis=None`: 전체 값을 하나로 펼쳐 계산함
- `axis=0`: 행 방향으로 내려가며 계산함, 열별 결과가 남음
- `axis=1`: 열 방향으로 이동하며 계산함, 행별 결과가 남음
- 핵심: axis는 계산 후 사라지는 축임

```python
arr = np.arange(1,7).reshape(2,3)
# [[1 2 3]
# [4 5 6]]

#최솟값 구하기
print("arr.min(axis=None):", arr.min(axis=None))

#axis = 0: 행 방향으로 내려가면서 같은 열의 최솟값 구하기
print("arr.min(axis=0):", arr.min(axis=0))  # [1 2 3]

#axis = 1: 열 방향으로 오른쪽으로 이동하면서 같은 행의 최솟값 구하기
print("arr.min(axis=0):", arr.min(axis=1))  # [1 4]
```

```python
# 합계
print("arr.sum():", arr.sum()) #axis 미작성 == 기본값(axis=None)
print("arr.sum(axis=0):", arr.sum(axis=0))
print("arr.sum(axis=1):", arr.sum(axis=1))
```

```python
# 평균(mean = 수학적으로 평균이라는 뜻)  ndarray.mean()

print("arr.mean():", arr.mean())
print("arr.mean(axis=0):", arr.mean(axis=0))
print("arr.mean(axis=1):", arr.mean(axis=1))
```
